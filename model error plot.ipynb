```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
errors = pd.read_csv('model_errors.csv',index_col=False)
errors.head()
```

```python
WR = errors.loc[(errors['position']=='WR') & (errors['type of error']=='MAE')]
RB = errors.loc[(errors['position']=='RB') & (errors['type of error']=='MAE')]
QB = errors.loc[(errors['position']=='QB') & (errors['type of error']=='MAE')]
```

```python
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
fig, axs = plt.subplots(3,1,figsize=(6,12))
weeks = range(1,11)
#plt.style.use('classic')
axs[0].plot(weeks,WR.loc[WR['model']=='RF'].iloc[:,3:].values[0], label="Random Forest")
axs[0].plot(weeks,WR.loc[WR['model']=='LR'].iloc[:,3:].values[0], label="Linear Regression")
axs[0].plot(weeks,WR.loc[WR['model']=='baseline'].iloc[:,3:].values[0], label="Naive Approach")
axs[0].set_ylabel("Mean Absolute Error")
axs[0].set_xlabel("Week")
axs[0].set_title('Wide Receiver')
axs[1].plot(weeks,RB.loc[RB['model']=='RF'].iloc[:,3:].values[0])
axs[1].plot(weeks,RB.loc[RB['model']=='LR'].iloc[:,3:].values[0])
axs[1].plot(weeks,RB.loc[RB['model']=='baseline'].iloc[:,3:].values[0])
axs[1].set_ylabel("Mean Absolute Error")
axs[1].set_xlabel("Week")
axs[1].set_title('Running Back')
axs[2].plot(weeks,QB.loc[QB['model']=='RF'].iloc[:,3:].values[0])
axs[2].plot(weeks,QB.loc[QB['model']=='LR'].iloc[:,3:].values[0])
axs[2].plot(weeks,QB.loc[QB['model']=='baseline'].iloc[:,3:].values[0])
axs[2].set_ylabel("Mean Absolute Error")
axs[2].set_xlabel("Week")
axs[2].set_title('Quarterback')
axs[0].legend()
plt.tight_layout()
```
